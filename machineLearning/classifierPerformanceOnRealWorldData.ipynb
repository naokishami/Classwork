In this hw assignment, we are looking at how different classifiers will perform on a real world dataset. First, we trained all of the classifiers on the original dataset. But then, we realized that the dataset consisted of an uneven amount of target values (too many no's in the dataset) so we resampled and retrained our classifiers.

***Question 2***

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from imblearn.over_sampling import RandomOverSampler 
```

```python
bank = pd.read_csv("./bank-additional/bank-additional.csv", sep=";")
bank
```

***Question 3a***

```python
bank = pd.get_dummies(bank, drop_first=True)
bank
```

***Question 3b***

```python
bank = bank.drop(columns = 'duration')
bank
```

```python
target = bank['y_yes']
bank = bank.drop(columns = 'y_yes')
bank
```

***Question 3c***

```python
cols_to_std = ['age', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 
                   'cons.conf.idx', 'euribor3m', 'nr.employed']

df_to_scale = bank[cols_to_std]

df_to_remain = bank.iloc[:,9:]

scaler = preprocessing.StandardScaler()
scaler = scaler.fit(df_to_scale)
scaled_df = scaler.transform(df_to_scale)

scaled_df = pd.DataFrame(scaled_df, columns = cols_to_std)

df = pd.concat([scaled_df, df_to_remain], axis = 1)
df
```

***Question 4***

```python
gnb = GaussianNB()
gnb.fit(df, target)
print(gnb.score(df, target))

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(df, target)
print(knn.score(df, target))

svm = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svm.fit(df, target)
print(svm.score(df, target))
```

***Question 8***

```python
ros = RandomOverSampler(random_state=(2021-4-22))
df_resampled, target_resampled = ros.fit_resample(df, target)
```

***Question 9***

```python
gnb = GaussianNB()
gnb.fit(df_resampled, target_resampled)
print(gnb.score(df_resampled, target_resampled))

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(df_resampled, target_resampled)
print(knn.score(df_resampled, target_resampled))

svm = make_pipeline(StandardScaler(), SVC(gamma='auto'))
svm.fit(df_resampled, target_resampled)
print(svm.score(df_resampled, target_resampled))
```

```python
confusion_matrix(target_resampled, gnb.predict(df_resampled))
```

```python
confusion_matrix(target_resampled, knn.predict(df_resampled))
```

```python
confusion_matrix(target_resampled, svm.predict(df_resampled))
```
